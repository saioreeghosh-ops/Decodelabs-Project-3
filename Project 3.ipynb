```python
import pandas as pd
import pandasql as ps

# Load the dataset
df = pd.read_excel("cleaned_data.xlsx")

# Define helper function
def run_query(q):
    return ps.sqldf(q, {"df": df})

```

```python
# Select specific columns to preview first 10 rows
run_query("SELECT OrderID, Product, TotalPrice FROM df LIMIT 10;")
```

```python
# Filter all orders that were shipped
run_query("SELECT * FROM df WHERE OrderStatus = 'Shipped';")

# Filter high-value orders (>= 2000)
run_query("SELECT OrderID, CustomerID, TotalPrice FROM df WHERE TotalPrice >= 2000;")
```

```python
# Count total orders per product
run_query("SELECT Product, COUNT(*) AS total_orders FROM df GROUP BY Product;")

# Count total orders per payment method
run_query("SELECT PaymentMethod, COUNT(*) AS total_orders FROM df GROUP BY PaymentMethod;")
```

```python
# Calculate total revenue and average order value
run_query("SELECT SUM(TotalPrice) AS total_sales, AVG(TotalPrice) AS avg_sales FROM df;")

# Average spend per product
run_query("SELECT Product, AVG(TotalPrice) AS avg_spend FROM df GROUP BY Product;")
```

```python
# Top 10 customers by total spend
run_query("""
SELECT CustomerID, SUM(TotalPrice) AS total_spent
FROM df
GROUP BY CustomerID
ORDER BY total_spent DESC
LIMIT 10;
""")
```

```python
#Business Insights Queries 
run_query("""
SELECT Product, SUM(TotalPrice) AS revenue
FROM df
GROUP BY Product
ORDER BY revenue DESC
LIMIT 5;
""")
```

```python
run_query("""
SELECT CouponCode, COUNT(*) AS usage_count
FROM df
GROUP BY CouponCode
ORDER BY usage_count DESC;
""")

```

```python
run_query("""
SELECT ReferralSource, SUM(TotalPrice) AS revenue
FROM df
GROUP BY ReferralSource
ORDER BY revenue DESC;
""")
```

```python
run_query("""
SELECT PaymentMethod, AVG(TotalPrice) AS avg_order_value
FROM df
GROUP BY PaymentMethod;
""")
```
